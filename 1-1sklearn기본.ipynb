```python
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

# 지도학습 - 회귀 
## 학습데이터 준비 (feature(X), target(y)

```python
# feature, 독립변수
x = np.arange(1,30,10).reshape(-1,1) 
# 1부터 30까지 10 차이나게 뽑음 
# 행은 알아서 열은 한개로 2차원 만들기 
x
```

```python
# target(label), 종속변수 
y = 2*x + 1
y
```

```python
sns.regplot(x=x, y=y)
```

## 학습기 객체 생성 

```python
model_lr = LinearRegression ()
model_lr
```

## 모델 학습하기
- 학습기 객체.fit()

```python
model_lr.fit(x,y)
```

## 테스트
- 예측 

```python
# x데이터를 넣어줘야함 -> 2차원 데이터였기때문에 2차원 넣어줘야함 
model_lr.predict([[40], [50]])
```

```python
# 결과 확인 
y=2*40+1
y
```
